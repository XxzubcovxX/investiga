```python
import requests
import pandas as pd

# 1. Definimos o endereço da API oficial da Câmara dos Deputados
url = "https://dadosabertos.camara.leg.br/api/v2/deputados"

# 2. Fazemos o pedido (request) dos dados
print("Baixando dados da Câmara...")
resposta = requests.get(url)

# 3. Pegamos a resposta em formato JSON e isolamos a parte dos 'dados'
dados_brutos = resposta.json()['dados']

# 4. Transformamos isso em uma tabela organizada (DataFrame) usando o Pandas
df_deputados = pd.DataFrame(dados_brutos)

# 5. Mostramos as 5 primeiras linhas da tabela
display(df_deputados.head())
```

```python
# 1. Vamos pegar o ID e o Nome do primeiro deputado da nossa tabela anterior (índice 0)
# Se quiser outro deputado, basta mudar o 0 para 1, 2, 3...
id_deputado = df_deputados.iloc[0]['id']
nome_deputado = df_deputados.iloc[0]['nome']

print(f"Buscando despesas do deputado(a): {nome_deputado} (ID: {id_deputado})")

# 2. Definimos o endereço da API para as despesas ESPECÍFICAS deste deputado
url_despesas = f"https://dadosabertos.camara.leg.br/api/v2/deputados/{id_deputado}/despesas"

# 3. Podemos passar "parâmetros" para a API para filtrar o que queremos.
# Vamos pedir os gastos do ano de 2023, ordenados do maior para o menor.
parametros = {
    "ano": 2023,
    "ordem": "DESC",
    "ordenarPor": "valorDocumento",
    "itens": 100 # Traz até 100 notas fiscais nesta chamada
}

# 4. Fazemos o pedido passando nossos filtros
print("Baixando as notas fiscais...")
resposta_despesas = requests.get(url_despesas, params=parametros)

# 5. Transformamos o resultado em uma nova tabela (DataFrame)
dados_despesas = resposta_despesas.json()['dados']
df_despesas = pd.DataFrame(dados_despesas)

# 6. Exibimos as primeiras linhas da tabela
display(df_despesas.head())
```
